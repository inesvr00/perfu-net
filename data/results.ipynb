```python
import os
import glob
import numpy as np
import matplotlib.pyplot as plt

from scipy.spatial.distance import cdist

# Directorio donde se guardaron las predicciones .npy
pred_dir = "/data/dev/perfu-net-1/wandb/run-20250407_170147-mlgov85f/files/predictions"
# Directorio donde est√°n las m√°scaras reales (GT)
mask_dir = "/data/dev/perfu-net-1/data/train/MASK"

# Coge todos los ficheros .npy de predicci√≥n
pred_files = sorted(glob.glob(os.path.join(pred_dir, "*.npy")))

print(f"Encontrados {len(pred_files)} ficheros de predicci√≥n en {pred_dir}")

def visualize_pred_and_mask(pred, gt, slice_name="", threshold=0.5):
    """
    Muestra la m√°scara GT y la predicci√≥n binarizada (> threshold)
    en un par de subplots para comparaci√≥n visual.
    """
    # Aseg√∫rate de que pred y gt tengan la misma shape en 2D
    if pred.ndim == 2 and gt.ndim == 2:
        pred_2d = pred
        gt_2d   = gt
    else:
        # Ejemplo si es 3D, toma primer canal/tiempo
        pred_2d = pred[..., 0]  # o usa pred[0] seg√∫n corresponda
        gt_2d   = gt[..., 0]

    pred_bin = (pred_2d > threshold).astype(np.uint8)

    fig, ax = plt.subplots(1, 2, figsize=(8, 4))
    ax[0].imshow(gt_2d, cmap='gray')
    ax[0].set_title(f"GT: {slice_name}")
    ax[0].axis("off")

    ax[1].imshow(pred_bin, cmap='gray')
    ax[1].set_title(f"Pred (> {threshold}): {slice_name}")
    ax[1].axis("off")

    plt.show()


def compute_metrics(pred, gt, threshold=0.5):
    pred_bin = (pred > threshold).astype(np.uint8)
    gt_bin   = (gt > 0.5).astype(np.uint8)
    
    sum_gt = np.sum(gt_bin)
    sum_pred = np.sum(pred_bin)

    # Caso especial: GT vac√≠a
    if sum_gt == 0:
        if sum_pred <=50:
            return {
                "Dice": 1.0,
                "Precision": 1.0,
                "Recall": 1.0,
                "IoU": 1.0,
                "Accuracy": 1.0,
                "F1": 1.0,
            }
        else:
            TN = np.sum(pred_bin == 0)
            total = pred_bin.size
            acc = TN / total if total > 0 else 0.0
            return {
                "Dice": 0.0,
                "Precision": 0.0,
                "Recall": 0.0,
                "IoU": 0.0,
                "Accuracy": acc,
                "F1": 0.0,
            }

    # Normal
    pred_flat = pred_bin.flatten()
    gt_flat   = gt_bin.flatten()
    
    TP = np.sum((pred_flat == 1) & (gt_flat == 1))
    FP = np.sum((pred_flat == 1) & (gt_flat == 0))
    FN = np.sum((pred_flat == 0) & (gt_flat == 1))
    TN = np.sum((pred_flat == 0) & (gt_flat == 0))
    
    dice = (2 * TP) / (2 * TP + FP + FN) if (2 * TP + FP + FN) != 0 else 0.0
    precision = TP / (TP + FP) if (TP + FP) != 0 else 0.0
    recall = TP / (TP + FN) if (TP + FN) != 0 else 0.0
    iou = TP / (TP + FP + FN) if (TP + FP + FN) != 0 else 0.0
    accuracy = (TP + TN) / (TP + TN + FP + FN) if (TP + TN + FP + FN) != 0 else 0.0
    f1 = dice

    return {
        "Dice": dice,
        "Precision": precision,
        "Recall": recall,
        "IoU": iou,
        "Accuracy": accuracy,
        "F1": f1,
    }


# Para acumular las m√©tricas de cada caso
all_metrics = []
# Para acumular solo casos donde la GT tiene p√≠xeles positivos (GT > 0)
all_metrics_non_empty = []

# Recorremos cada fichero de predicci√≥n y buscamos su m√°scara correspondiente
for pf in pred_files:
    pred_filename = os.path.basename(pf)  # Ej: "case_01_10.npy"
    mask_path = os.path.join(mask_dir, pred_filename)  # asumimos mismo nombre

    if not os.path.exists(mask_path):
        print(f"No existe la m√°scara para {pred_filename} en {mask_dir}. Se omite.")
        continue

    # Cargamos la predicci√≥n y la m√°scara
    pred_array = np.load(pf)
    gt_array   = np.load(mask_path)

    # (Opcional) Visualizaci√≥n
    visualize_pred_and_mask(pred_array, gt_array, slice_name=pred_filename, threshold=0.5)
    
    # Calculamos las m√©tricas
    metrics = compute_metrics(pred_array, gt_array, threshold=0.5)
    print(f"M√©tricas para {pred_filename}:")
    for key, value in metrics.items():
        # Si es infinito, lo mostramos de forma clara
        if value == np.inf:
            print(f"  {key}: Inf")
        else:
            print(f"  {key}: {value:.4f}")
    
    # A√±adimos a la lista global
    all_metrics.append(metrics)
    
    # Si la GT tiene p√≠xeles positivos, a√±adimos a la lista de no vac√≠os
    if np.sum(gt_array > 0.5) > 0:
        all_metrics_non_empty.append(metrics)

# C√°lculo de la media de m√©tricas (todos los casos)
if all_metrics:
    mean_metrics_all = {}
    keys_metrics = list(all_metrics[0].keys())
    for key in keys_metrics:
        valid_vals = [m[key] for m in all_metrics if np.isfinite(m[key])]
        if len(valid_vals) == 0:
            mean_metrics_all[key] = np.inf
        else:
            mean_metrics_all[key] = np.mean(valid_vals)

    print("\nMedia de m√©tricas en TODOS los casos (GT vac√≠a y GT no vac√≠a):")
    for key, value in mean_metrics_all.items():
        if value == np.inf:
            print(f"  {key}: Inf")
        else:
            print(f"  {key}: {value:.4f}")
else:
    print("No se han procesado m√©tricas para ning√∫n caso.")

# C√°lculo de la media de m√©tricas (solo casos GT>0)
if all_metrics_non_empty:
    mean_metrics_non_empty = {}
    keys_metrics = list(all_metrics_non_empty[0].keys())
    for key in keys_metrics:
        valid_vals = [m[key] for m in all_metrics_non_empty if np.isfinite(m[key])]
        if len(valid_vals) == 0:
            mean_metrics_non_empty[key] = np.inf
        else:
            mean_metrics_non_empty[key] = np.mean(valid_vals)

    print("\nMedia de m√©tricas SOLO en casos donde la GT tiene p√≠xeles positivos (GT>0):")
    for key, value in mean_metrics_non_empty.items():
        if value == np.inf:
            print(f"  {key}: Inf")
        else:
            print(f"  {key}: {value:.4f}")
else:
    print("No hay casos con GT>0. No se puede calcular la media de segmentaci√≥n real.")
```

```python
import os
import glob
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
import torch

def dice_coef(pred, gt, smooth=1e-5):
    """
    Calcula el coeficiente de Dice cl√°sico.
    """
    pred_flat = pred.flatten()
    gt_flat = gt.flatten()
    intersection = np.sum(pred_flat * gt_flat)
    return (2 * intersection + smooth) / (np.sum(pred_flat) + np.sum(gt_flat) + smooth)


def precision_score(pred, gt, smooth=1e-5):
    """
    Calcula la precisi√≥n (precision) cl√°sica.
    """
    pred_flat = pred.flatten()
    gt_flat = gt.flatten()
    true_positive = np.sum(pred_flat * gt_flat)
    return (true_positive + smooth) / (np.sum(pred_flat) + smooth)


def recall_score(pred, gt, smooth=1e-5):
    """
    Calcula el recall (sensibilidad) cl√°sico.
    """
    pred_flat = pred.flatten()
    gt_flat = gt.flatten()
    true_positive = np.sum(pred_flat * gt_flat)
    return (true_positive + smooth) / (np.sum(gt_flat) + smooth)

# Directorios de datos
pred_dir = "/data/dev/perfu-net-1/wandb/run-20250408_114354-v872ozbj/files/predictions"
mask_dir = "/data/dev/perfu-net/data/train/MASK"
ctp_dir  = "/data/dev/perfu-net/data/train/CTP"

# Listado de ficheros .npy
pred_files = sorted(glob.glob(os.path.join(pred_dir, "*.npy")))
mask_files = sorted(glob.glob(os.path.join(mask_dir, "*.npy")))
ctp_files  = sorted(glob.glob(os.path.join(ctp_dir, "*.npy")))

# Casos comunes
common = sorted(set(os.path.basename(f) for f in pred_files)
                .intersection(os.path.basename(f) for f in mask_files)
                .intersection(os.path.basename(f) for f in ctp_files))
print(f"Casos encontrados: {len(common)}")

# Acumuladores
dice_all = []
precision_all = []
recall_all = []
dice_non_empty = []
precision_non_empty = []
recall_non_empty = []

for name in common:
    # Carga
    pred = np.squeeze(np.load(os.path.join(pred_dir, name)))
    mask = np.squeeze(np.load(os.path.join(mask_dir, name)))
    brain = np.load(os.path.join(ctp_dir, name))
    brain_slice = brain[1, 1]

    # Binarizaci√≥n al umbral 0.5
    pred_bin = (pred > 0.5).astype(np.uint8)
    mask_bin = (mask > 0.5).astype(np.uint8)

    # C√°lculo de m√©tricas
    dice_value = dice_coef(pred_bin, mask_bin)
    prec_value = precision_score(pred_bin, mask_bin)
    rec_value = recall_score(pred_bin, mask_bin)

    dice_all.append(dice_value)
    precision_all.append(prec_value)
    recall_all.append(rec_value)

    if mask_bin.sum() > 0:
        dice_non_empty.append(dice_value)
        precision_non_empty.append(prec_value)
        recall_non_empty.append(rec_value)

    # Visualizaci√≥n
    plt.figure(figsize=(8, 8))
    plt.imshow(brain_slice, cmap="gray", interpolation='none')
    plt.contour(mask_bin, colors='blue', linewidths=2)
    plt.contour(pred_bin, colors='red', linewidths=2)

    # Leyenda
    legend = [
        Patch(facecolor='gray', edgecolor='gray', label='Imagen Cerebral'),
        Patch(facecolor='blue', edgecolor='blue', label='GT'),
        Patch(facecolor='red', edgecolor='red', label='Predicci√≥n')
    ]
    plt.legend(handles=legend, loc='upper right')

    # Texto con m√©tricas en la imagen
    metrics_text = f"Dice: {dice_value:.3f}\nPrec: {prec_value:.3f}\nRec: {rec_value:.3f}"
    plt.gca().text(0.20, 0.70, metrics_text,
                   verticalalignment='top', color='yellow', fontsize=12,
                   )

    plt.title(f"Caso: {name}")
    plt.axis('off')
    plt.show()

# Medias globales
def print_stats(name, values):
    print(f"{name}: media={np.mean(values):.3f}, std={np.std(values):.3f}, muestras={len(values)}")

print_stats("Dice (todos)", dice_all)
print_stats("Precision (todos)", precision_all)
print_stats("Recall (todos)", recall_all)

if dice_non_empty:
    print_stats("Dice (GT no vac√≠a)", dice_non_empty)
    print_stats("Precision (GT no vac√≠a)", precision_non_empty)
    print_stats("Recall (GT no vac√≠a)", recall_non_empty)
else:
    print("No hay casos con GT no vac√≠a para estad√≠sticas.")
```
